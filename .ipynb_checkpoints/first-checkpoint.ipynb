#MatplotLib Basics#

```python
height = [150,170,165,175,185]
weight = [70,80,80,85,90]

plt.plot(height,weight)
```

```python
plt.plot(height,weight)
plt.title("Relation between height and weight")
plt.xlabel("height")
plt.ylabel("weight")
```

```python
Cal_burnt = [65,75,85,90,99]
plt.plot(Cal_burnt)
plt.plot(weight)
```

```python
plt.plot(Cal_burnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','weight'],loc ='upper right')
```

```python
plt.figure(figsize=(15,5))
plt.plot(Cal_burnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','weight'],loc ='upper right')
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# plot on 0 row and 0 column
ax[0,0].plot(Cal_burnt, 'go')
# plot on 0 row and 1 column
ax[0,1].plot(weight)
# set titles for subplots
ax[0,0].set_title("Calories Burnt")
ax[0,1].set_title("Weight")
# set ticks for each of these persons
ax[0,0].set_xticks(ticks=[0,1,2,3]);
ax[0,1].set_xticks(ticks=[0,1,2,3]);
# set labels for each of these persons
ax[0,0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
ax[0,1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
```

```python

x=np.array([1, 2, 3, 4, 5])
# making subplots
fig, ax = plt.subplots(2, 2)
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)
plt.show()
```

#Working on 3D picture#

```python
fig = plt.figure()
axis = plt.axes(projection='3d')
```

```python
import numpy as np
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)
ax.plot3D(x,y,z,'red')
ax.set_title('3d line plot')
plt.show()
```

```python

ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)
c = x + y
ax.scatter(x,y,z,c=c)
ax.set_title('3d scatter plot')
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,10),np.ones(10))
y = x.copy().T
z = np.cos(x**2 + y**3)

fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='Spectral',
                edgecolor='blue')
ax.set_title('surface plot')
plt.show()
```

```python

def function(x, y):
 return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
X, Y = np.meshgrid(x, y)
Z = function(X, Y)
fig = plt.figure(figsize=(15, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='magma', alpha=0.8)
ax.set_title('3D Contour Plot of function(x, y) =\
 sin(sqrt(x^2 + y^2))', fontsize=20)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
plt.show()
```

###Importing dataset###

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Flipkart_Mobiles.csv')
df = df.dropna()
df.head(5)
```

```python
phone_by_rating = df.groupby(['Brand', 'Model'])['Rating'].mean().nlargest(8)
phone_by_rating
```

###Line chart###

```python
phone_by_rating = df.groupby(['Brand', 'Model']).Rating.mean().nlargest(7)
x = phone_by_rating.index.tolist()
x = [f"{brand} - {model}" for brand, model in x] 
y = phone_by_rating.values.tolist()

plt.figure(figsize=(10,6))
plt.title('Mean Rating of each mobile')
plt.xlabel('Brand - Model')
plt.ylabel('Mean Ratings')
plt.xticks(rotation=45, ha='right')

plt.bar(x, y)

plt.show()
```

###Histogram###



```python
plt.title('moblie MRP (price) distribution')
plt.xlabel('Moblie_MRP')
plt.ylabel('Frequency')
plt.hist(df['Selling Price'], bins=20, color='lightblue');
```

###Box Plots Distributions of Sales###

```python
data = df[['Selling Price']]
red_diamond = dict(markerfacecolor='r', marker='D')

plt.title('Mobile sales distribution')
plt.boxplot(data.values,labels=['Moblie Sales'], flierprops=red_diamond);
```
