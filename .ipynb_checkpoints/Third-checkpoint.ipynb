# One Sample T-test#


```python
import numpy as np
ages = np.array([45,89,23,46,12,69,45,24,34,67])
pop_mean = 50

t_statistic, p_value = ttest_ind(ages, pop_mean)
print("t-stastic", t_statistic)
print("p-value", p_value)
```

#Two Sample T-test#

```python
from scipy.stats import ttest_ind
sample1 = np.array([29,30,33,41,38,36,35,31,29,30])
sample2 = np.array([31,32,33,39,30,33,30,28,29,31])

t_statistic, p_value = ttest_ind(sample1, sample2)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

#Paried T-test#

```python
from scipy.stats import ttest_rel
before = np.array([5,8,12,15,18])
after = np.array([7,10,13,16,19])

t_statistic, p_value = ttest_rel(before,after)
print ("t-statistic:",t_statistic)
print ("p-value:",p_value)
```

#Annova Test F-test#

```python
import statsmodels.formula.api as smf
import pandas as pd
import statsmodels.api as sm

data = pd.DataFrame({'Group':['a','a','a','b','b','b','c','c','c','d','d','d'],
                     'value':[20,19,24,23,27,15,25,30,14.5,30,27,18]})

model = smf.ols('value ~ C(Group)',data=data).fit()
aov_table = sm.stats.anova_lm(model,type=2)
print(aov_table)
```

#Linear Regression#

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

model = LinearRegression()


model.fit(x, y)

print("Coefficients:", model.coef_, model.intercept_)

y_pred = model.predict(x)

plt.scatter(x, y, color = 'blue', label="actual data")
plt.plot(x, y_pred, color = 'green', label='regression line')
plt.xlabel('Independent var (x)')
plt.ylabel('Dependent var (y)')
plt.title('Linear Regression')
```
